```python
import pandas as pd
import re
```

```python
df = pd.read_excel("Filebaru.xlsx")
df.head()
```

```python
df2 = df.drop(columns=['No' , 'Nama Ibu/Suami' , 'Alamat' , 'HPHT' , 'TP'] , axis=-1)
df2.head()
```

```python
df2.isna().sum()
```

```python
jdl = ['Umur' , 'BB' , 'TB' , 'UK' , 'LILA']
for j in jdl :
    df2[j] = df2[j].apply(lambda x : float(eval(re.sub(r'[^\d.]', '', str(x).replace(',' , '.')))))

df2.to_csv('terpakai.csv')
```
